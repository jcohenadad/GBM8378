Pour un scan donné, l’onde acoustique se propage selon l’axe des z, lorsque l’onde rencontre un changement d’indice, un signal est réfléchi et pourra être mesuré par le transducteur. Il est à noter que le signal mesuré est un signal temporel et que le signal recherché est une mesure spatiale.<br><br>

Déconvolution :
Les données recueillies par chaque transducteur peuvent être représentées mathématiquement approximativement<sup>1</sup> comme :

$g(t)=\int_{-\infty}^{\infty} h(t-\tau)\cdot f(\tau) d\tau = h(\tau)\star f(\tau)$
(1)<br>
où :<ul>
    $g(t)=$données recueillies<br>
    $f(t)=$données réelles<br>
    $h(t)=$réponse temporelle du transducteur<br>
    </ul>
</b>
<sup>1</sup>Le signal provenant de réflexions sur des éléments de d’autres transducteur n’est pas considéré.

```python
## Title: Ultrasound Forward Simulator
## Authors: Jonathan Poree, Samuel Desmarais --- Provost Ultrasound Lab
## Inspired from work in: Section 2.3 of https://doi.org/10.1088/1361-6560/aae3c3

from labUSfunctions import ussimforward, prepare_animation

##################
### PARAMETERS ###
##################

## PROBE
prm = {}
prm['pitch'] = 0.5*1e-3  # Spacing between the center of 2 elements (m)
prm['width'] = 0.8117*1540/(14e6)  # Width of element (m)
prm['height'] = 1.5*1e-3  # Height of element (m)
prm['nele'] = 1  # Number of elements
prm['BW'] = .5  # 50% Bandwidth
prm['eleBW'] = 1  # 100% Bandwidth
prm['hardbaffle'] = True  # Use hard baffle =True or soft baffle =False
prm['fc'] = 14e6  # Central frequency of the transducer (Hz)

## MEDIUM
prm['c'] = 1540  # Propagation velocity (m/s)
prm['lbd'] = prm['c']/prm['fc']  # Wavelenght (m)

## IMAGING SEQUENCE
prm['fs'] = 4 * prm['fc']  # Sampling frequency of the system (Hz)
prm['elevfocus'] = 64 * prm['lbd']
prm['fspulse'] = 10*prm['fs']  # Sampling frequency for the pulse (Hz)
prm['pulse'] = np.sin(2*np.pi*prm['fc']*np.arange(0, 2/prm['BW']/prm['fc']+1/prm['fspulse'], 1/prm['fspulse']))
prm['pulse'] = prm['pulse'] * np.hanning(prm['pulse'].size)  # Pulse Shape
prm['planewaveangle'] = 0  # Plane wave emission angle (degrees)

## SIMULATION
prm['useprobeheight'] = False  # Take probe height into account =True or not =False
prm['gridsize'] = 256  # Number of samples per side of the grid
prm['fovx'] = prm['pitch'] * 25.6  # Lateral field of view (m)
prm['fovz'] = prm['lbd'] * 128  # Axial field of view (m)

#############
### SETUP ###
#############

## Grid Definition
zgrid, xgrid = np.meshgrid(np.linspace(0, 1, prm['gridsize'])*prm['fovz'], np.linspace(-0.5, 0.5, prm['gridsize'])*prm['fovx'])
ygrid = np.zeros_like(xgrid)

## Delays Definition
delays = [0.]    # Firing time of the different transductors

## Apodization Definition
apod = np.ones_like(delays)

##################
### SIMULATION ###
##################
RF = ussimforward(xgrid, ygrid, zgrid, delays, apod, prm)

######################
### OUTPUT US WAVE ###
######################
anim = prepare_animation(RF, prm['fovx'], prm['fovz'])
display(HTML(anim.to_html5_video()))
```

```python
##################
### PARAMETERS ###
##################

## PROBE
prm = {}
prm['pitch'] = 0.5*1e-3  # Spacing between the center of 2 elements (m)
prm['width'] = 0.8117*1540/(14e6)  # Width of element (m)
prm['height'] = 1.5*1e-3  # Height of element (m)
prm['nele'] = 1  # Number of elements
prm['BW'] = .5  # 50% Bandwidth
prm['eleBW'] = 1  # 100% Bandwidth
prm['hardbaffle'] = True  # Use hard baffle =True or soft baffle =False
prm['fc'] = 14e6  # Central frequency of the transducer (Hz)

## MEDIUM
prm['c'] = 1540  # Propagation velocity (m/s)
prm['lbd'] = prm['c']/prm['fc']  # Wavelenght (m)

## IMAGING SEQUENCE
prm['fs'] = 4 * prm['fc']  # Sampling frequency of the system (Hz)
prm['elevfocus'] = 64 * prm['lbd']
prm['fspulse'] = 10*prm['fs']  # Sampling frequency for the pulse (Hz)
prm['pulse'] = np.sin(2*np.pi*prm['fc']*np.arange(0, 2/prm['BW']/prm['fc']+1/prm['fspulse'], 1/prm['fspulse']))
prm['pulse'] = prm['pulse'] * np.hanning(prm['pulse'].size)  # Pulse Shape
prm['planewaveangle'] = 0  # Plane wave emission angle (degrees)

## SIMULATION
prm['useprobeheight'] = False  # Take probe height into account =True or not =False
prm['gridsize'] = 256  # Number of samples per side of the grid
prm['fovx'] = prm['pitch'] * 25.6  # Lateral field of view (m)
prm['fovz'] = prm['lbd'] * 128  # Axial field of view (m)

#############
### SETUP ###
#############

## Grid Definition
zgrid, xgrid = np.meshgrid(np.linspace(0, 1, prm['gridsize'])*prm['fovz'], np.linspace(-0.5, 0.5, prm['gridsize'])*prm['fovx'])
ygrid = np.zeros_like(xgrid)

## Delays Definition
delays = [0.]    # Firing time of the different transductors

## Apodization Definition
apod = np.ones_like(delays)

##################
### SIMULATION ###
##################
RF = ussimforward(xgrid, ygrid, zgrid, delays, apod, prm)

######################
### OUTPUT US WAVE ###
######################
anim = prepare_animation(RF, prm['fovx'], prm['fovz'])
display(HTML(anim.to_html5_video()))
```
